```python
import numpy as np
import matplotlib.pyplot as plt
```

## All Tess

```python
# importing the data into arrays
# first column of the data file is phase data
TessPhase = np.loadtxt("GD356allTess.dat", usecols=(0))
# second column of the data file is flux data
TessFlux = np.loadtxt("GD356allTess.dat", usecols=(1))

# creating a figure
plt.figure()
# plotting the data as black dots with no line connecting them
plt.plot(TessPhase, TessFlux, color='black', marker='.', linestyle='none', markersize=2)
# adding the x and y axis labels
plt.xlabel('Phase')
plt.ylabel('Demeaned Relative Flux')
# get y-axis limits of the plot
low, high = plt.ylim()
# find the new limits
bound = max(abs(low), abs(high))
# set new limits
plt.ylim(-bound, bound)
# plot a line through y=0
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```
