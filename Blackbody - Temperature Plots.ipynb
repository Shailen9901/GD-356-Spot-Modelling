```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wav, T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity

wavelengths = np.linspace(1e-7, 1e-6, 1000) 

# Black Body for GD 356
# T = 7510K
intensityGD356 = planck(wavelengths, 7510.)

plt.figure()
figure(figsize=(12, 9), dpi=80)
plt.plot(wavelengths*1e9, intensityGD356, 'r-') 
plt.show()
```

# TESS - Medium Radius

```python
contrast = np.array([0.10949, 0.15681, 0.22019, 0.35688])
wavelengths2 = np.array([788.7781855, 615.8435393, 471.8872246, 356.5647932])*1e-9

contrast = 1 - contrast

inten = planck(wavelengths2, 7510.)

intensityTESS = [0, 0, 0, 0]

for i in range(4):
    intensityTESS[i] = inten[i] * contrast[i]
```

```python
TESS_test = planck(wavelengths, 7100)

plt.figure()
figure(figsize=(12, 9), dpi=80)
plt.plot(wavelengths*1e9, intensityGD356, 'r-', label='Star') 
plt.plot(wavelengths*1e9, TESS_test, 'g-', label='Test') 
plt.plot(wavelengths2*1e9, intensityTESS, 'b-', label='Spot')
plt.legend()
plt.show()
```
