```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
h = 6.626e-34
c = 3.0e+8
k = 1.38e-23

def planck(wav, T):
    a = 2.0*h*c**2
    b = h*c/(wav*k*T)
    intensity = a/ ( (wav**5) * (np.exp(b) - 1.0) )
    return intensity

wavelengths = np.linspace(1e-7, 1e-6, 1000) 

# Black Body for GD 356
# T = 7510K
intensityGD356 = planck(wavelengths, 7510.)

plt.figure()
figure(figsize=(12, 9), dpi=80)
plt.plot(wavelengths*1e9, intensityGD356, 'r-') 
plt.show()
```

# Wavelengths

```python
# TESS = 788.7781855
# ZTF-r = 646.375
# LT = 615.8435393
# ZTF-g = 482.950
# PTO = 471.8872246
# WHT = 356.5647932
# NUV = 244.60698626444548
# NUV (2550 - 2950) = 278.9703445161371
# NUV (2150 - 2550) = 235.8028267307681
# NUV (1630 - 2150) = 195.76162964730503
# FUV = 135.89896930381835
# FUV (1520 - 1715) = 159.3242206252428
# FUV (1330 - 1520) = 141.1094441960065
# FUV (1120 - 1330) = 125.81681569145482


wavelengths2 = np.array([788.7781855, 646.375, 615.8435393, 482.950, 471.8872246, 356.5647932, 244.60698626444548,\
                        278.9703445161371, 235.8028267307681, 195.76162964730503, 135.89896930381835,\
                        159.3242206252428, 141.1094441960065, 125.81681569145482])*1e-9
```

# 30 Degrees

```python
contrast = np.array([0.04618, 0.05509, 0.06984, 0.08973, 0.10118, 0.12773, 0.23164, 0.21885, 0.28845, 0.20529, 0.73682,\
                     0.49299, 0.85734, 0.99324])

contrast = 1 - contrast

# contrast = (F_star - F_spot)/F_star = 1 - F_spot/F_star 
# 1 - contrast = F_spot/F_star

inten = planck(wavelengths2, 7510.)

intensitySpot = [0] * 14

for i in range(14):
    intensitySpot[i] = inten[i] * contrast[i]
```

```python
plt.figure()
figure(figsize=(12, 9), dpi=80)
plt.plot(wavelengths*1e9, intensityGD356, 'r-', label='Star') 
plt.plot(wavelengths2*1e9, intensitySpot, 'b.', label='Spot')
plt.legend()
plt.show()
```

```python
temperatures = np.linspace(7210, 7510, 301, endpoint=True)

min_res = (10e100)

for temp in temperatures:
    
    Spot_test = planck(wavelengths2, temp)
    res = np.sum((Spot_test - intensitySpot)**2)
    
    if res <= min_res:
        min_res = res
        best_temp = temp
```

```python
min_res
```

```python
best_temp
```

```python
Spot_30 = planck(wavelengths, best_temp)

plt.figure()
figure(figsize=(12, 9), dpi=80)
plt.plot(wavelengths*1e9, intensityGD356, 'r-', label='Star') 
plt.plot(wavelengths*1e9, Spot_30, 'g-', label='7294K') 
plt.plot(wavelengths2*1e9, intensitySpot, 'b.', label='Spot')
plt.legend(fontsize="15")
plt.show()
```
