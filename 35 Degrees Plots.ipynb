```python
import numpy as np
import starry
import matplotlib.pyplot as plt


starry.config.lazy = False
starry.config.quiet = True
```

```python
meanPhase = np.linspace(0, 2, num=200, endpoint=True)
theta = np.multiply(meanPhase, 360)
```

```python
map = starry.Map(ydeg=30)
map.spot(contrast=0.04598, radius=35, lat=50.64638, lon=45)
map.inc = 39.35362

TESS_newFlux = map.flux(theta=theta)/np.mean(map.flux(theta=theta))
```

```python
plt.figure(figsize=(12, 5))

plt.title('NUV - 30 Degree Radius Model', fontsize=20)
plt.plot(meanPhase, NUV_newFlux, label='NUV',linestyle='--')
plt.plot(meanPhase, NUV3_newFlux, label='NUV (2550 - 2950)')
plt.plot(meanPhase, NUV2_newFlux, label='NUV (2150 - 2550)')
plt.plot(meanPhase, NUV1_newFlux, label='NUV (1630 - 2150)')
plt.xlabel("Rotational phase", fontsize=20)
plt.legend()
plt.ylabel("Flux", fontsize=20)
plt.xlabel('Rotational Phase')
plt.ylabel('Flux')
plt.axhline(y=1, color='r', linestyle='--')
plt.ylim(np.amin(NUV2_newFlux), np.amax(NUV2_newFlux));
```
