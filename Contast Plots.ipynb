```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# TESS - Large Radius

```python
contrast = [0.04139, 0.05921, 0.08305, 0.13288]
wavelength = [788.7781855, 615.8435393, 471.8872246, 356.5647932]

# creating a figure
plt.figure()
figure(figsize=(8, 6), dpi=80)
plt.title('TESS Large Radius Model', fontsize=20)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength, contrast, color='black', marker='.', linestyle='-', markersize=5)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)', fontsize=20)
plt.ylabel(r'Contrast $\frac{L_{spot}}{L_{star}}$', fontsize=20)
plt.show()
```
