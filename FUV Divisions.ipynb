```python
import numpy as np
import matplotlib.pyplot as plt

# importing the data into arrays
# first column of the data file is wavelength data
wavelength = (np.loadtxt("HST_STIS_FUV.25MAMA_G140L.dat", usecols=(0)))/10
# second column of the data file is transmission data
transmission = np.loadtxt("HST_STIS_FUV.25MAMA_G140L.dat", usecols=(1))
```

```python
difference_array = np.absolute(wavelength-114.76)
index1 = difference_array.argmin()

difference_array = np.absolute(wavelength-133.25)
index2 = difference_array.argmin()

wavelength12 = wavelength[index1:index2+1]
transmission12 = transmission[index1:index2+1]
index2
```

```python
from matplotlib.pyplot import figure

# creating a figure
plt.figure()
figure(figsize=(8, 6), dpi=80)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength12, transmission12, color='black', marker='.', linestyle='none', markersize=2)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Sensitivity')
plt.show()

plt.figure()
figure(figsize=(8, 6), dpi=80)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength12, transmission12, color='black', marker='.', linestyle='-', markersize=0)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Sensitivity')
plt.show()
```

```python
mean = np.sum(wavelength12 * transmission12) / np.sum(transmission12)
mean
```
