```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
phase = [45.33061, 48.87513, 51.61675, 68.32525]
wavelength = [788.7781855, 615.8435393, 471.8872246, 356.5647932]

# creating a figure
plt.figure()
figure(figsize=(8, 6), dpi=80)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength, phase, color='black', marker='.', linestyle='-', markersize=5)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)', fontsize=20)
plt.ylabel(r'Phase (Longitude of Spot)', fontsize=20)
plt.show()
```
