```python
import numpy as np
import matplotlib.pyplot as plt

# importing the data into arrays
# first column of the data file is wavelength data
wavelength = np.loadtxt("LT_RISE_V+R_filter_throughput.csv", usecols=(0), delimiter=',')
# second column of the data file is transmission data
transmission = np.loadtxt("LT_RISE_V+R_filter_throughput.csv", usecols=(1), delimiter=',')
```

```python
from matplotlib.pyplot import figure

# creating a figure
plt.figure()
figure(figsize=(8, 6), dpi=80)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength, transmission, color='black', marker='.', linestyle='none', markersize=2)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Throughput (%)')
plt.show()

plt.figure()
figure(figsize=(8, 6), dpi=80)
# plotting the data as black dots with no line connecting them
plt.plot(wavelength, transmission, color='black', marker='.', linestyle='-', markersize=0)
# adding the x and y axis labels
plt.xlabel('Wavelength (nm)')
plt.ylabel('Throughput (%)')
plt.show()
```

```python
mean = np.sum(wavelength * transmission) / np.sum(transmission)
mean
```
